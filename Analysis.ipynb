<h1> Job  Analysis </h1>

## Objective
The objective of this project is to analyze job postings data and identify trends in:

- Job categories
- Hiring locations
- Companies hiring the most candidates
- Job types
- Hiring trends over time
- Most demanded skills

The analysis is performed using:
- Pandas for data manipulation
- Matplotlib for data visualization

# Importing Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datasets
```

# Loading The Dataset

```python
data = datasets.load_dataset("azrai99/job-dataset")
```

```python
job  = data['train'].to_pandas()
job.head()
```

# Dataset Overview

```python
job.info()
```

# Question 1

## Which job categories have the highest number of job openings?

This analysis helps identify industries with the greatest hiring demand.

```python
top_categories = job['category'].value_counts().head(10)
plt.figure(figsize=(15,8))
plt.bar(top_categories.index, top_categories.values)
plt.xticks(rotation=45, ha='right')
plt.title("Top 10 Job Categories")
plt.xlabel("Category")
plt.ylabel("Number of Jobs")
plt.show()
```

# Question 2

## Which locations have the highest number of job opportunities?

This helps identify major employment hubs.

```python
top_locations = job['location'].value_counts().head(10)

plt.figure(figsize=(12,6))
plt.bar(top_locations.index, top_locations.values)
plt.xticks(rotation=45, ha='right')
plt.title("Top Hiring Locations")
plt.xlabel("Location")
plt.ylabel("Number of Jobs")

plt.show()
```

# Question 3

## What is the distribution of job types?

This analysis shows the proportion of Full-Time, Part-Time, Contract, and Internship positions.

```python
job_types = job['type'].value_counts()

plt.figure(figsize=(85,5))

plt.pie(
    job_types,
    labels=job_types.index,
    autopct='%1.1f%%',
)


plt.title("Job Type Distribution")

plt.show()
```

# Question 4

## Which companies are posting the most job opportunities?

```python

top_companies = job['company'].value_counts().head(15)
plt.figure(figsize=(10,5))
plt.barh(top_companies.index, top_companies.values)
plt.title("Top Hiring Companies")
plt.xlabel("Company")
plt.ylabel("Number of Jobs")

plt.show()
```

# Question 5

## Which job titles appear most frequently?

This helps identify roles that are currently in high demand.

```python
top_jobs = job['job_title'].value_counts().head(20)
plt.figure(figsize=(10,5))
top_jobs.sort_values(ascending= True)
plt.barh(top_jobs.index, top_jobs.values)
plt.title("Most Common Job Titles")
plt.xlabel("Number of Jobs")
plt.show()
```

# Question 6

## How has hiring activity changed over time?

```python
job['listingDate'] = pd.to_datetime(job['listingDate'])
monthly_jobs = job.groupby(
    job['listingDate'].dt.to_period('M')
).size()
plt.figure(figsize=(10,5))
monthly_jobs.plot()
plt.title("Monthly Hiring Trend")
plt.xlabel("Month")
plt.ylabel("Number of Jobs")
plt.show()
```
